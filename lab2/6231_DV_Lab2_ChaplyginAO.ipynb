### Что делает geom_col()? Чем он отличается от geom_bar()?
geom_bar() - высота столбцов завист от количества элеметов с таким значнием</br>
geom_col() - высота столбцов отражает зависимость элементов между двумя атрибутами

```python
ggplot(data=diamonds) +\
geom_col(mapping=aes(x="cut", y="carat"))
```

### Что произойдёт, если 
'ggplot(data=diamonds) +\
'geom_bar(mapping=aes(x="cut", y="..prop.."), group=1)
или
ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="cut", fill="color", y="..prop.."))
### запустить без "group=1"? Почему?

Графики потряют смысл из за искажения данных для отображения

```python
ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="cut", y="..prop.."), group=1)
```

```python
ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="cut", y="..prop.."))
```

### Зачем тут coord_fixed()? Что делает geom_abline()?
